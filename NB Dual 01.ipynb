```python
import pandas as pd
```

```python
import sfa, bqa, tools
```

```python
from importlib import reload
```

```python
test = "ds"
scale = 1
cid = "01"
stream_n = 1
desc = "dev_test_02"
seq_id = "t2"
verbose = False
verbose_query = True
qual = None
save = True
verbose_iter = False

timestamp = pd.Timestamp.now()  # "UTC"

seq = tools.tpc_stream(test=test, n=stream_n)
#seq = seq[0:5]

test_name = tools.make_name(db="bqsf", test=test, cid=cid, kind="results",
                            datasource="ds_1GB_01", desc=desc, ext="", timestamp=timestamp)
```

```python
sf = sfa.SFTPC(test=test,
               scale=scale,
               cid=cid,
               warehouse="TEST9000",
               desc=desc,
               verbose=verbose,
               verbose_query=verbose_query)
sf.timestamp = timestamp
sf.results_dir = test_name
sf.connect()
sf.query_seq(seq=seq, seq_id=seq_id, qual=qual, save=save, verbose_iter=verbose_iter)
sf.close()
```

```python
bq = bqa.BQTPC(test=test,
               scale=scale,
               cid=cid,
               desc=desc,
               verbose_query=verbose_query, verbose=verbose)
bq.timestamp = timestamp
bq.results_dir = test_name
bq.query_seq(seq, seq_id=seq_id, qual=qual, save=save, verbose_iter=verbose_iter)
```
